```python
#导入数据
import pandas as pd
df1 = pd.read_excel(r'D:\pythonProject\a.xlsx') 
df2 = pd.read_excel(r'D:\pythonProject\t.xlsx')
df3 = pd.read_excel(r'D:\pythonProject\k.xlsx')
df1.columns = ["Timestamp", "Accelaration-X", "Accelaration-Y", "Accelaration-Z"] 
df2.columns = ["Timestamp", "AngularVelocity-X", "AngularVelocity-Y", "AngularVelocity-Z"]
df3.columns = ["Timestamp","latitude", "longtitude", "altitude", "speed", "course", "hacc"]
#每秒钟取均值
from datetime  import datetime
result = pd.DataFrame()
df1['minute'] = df1['Timestamp'].apply(lambda x : datetime.strftime(x ,'%Y-%m-%d %H:%M:%S'))
data1 = df1.groupby('minute').mean().reset_index()
df2['minute'] = df2['Timestamp'].apply(lambda x : datetime.strftime(x ,'%Y-%m-%d %H:%M:%S'))
data2 = df1.groupby('minute').mean().reset_index()
df3['minute'] = df3['Timestamp'].apply(lambda x : datetime.strftime(x ,'%Y-%m-%d %H:%M:%S'))
data3 = df3.groupby('minute').mean().reset_index()
#三张表连一起
result =pd.merge(pd.merge(data1,data2,on = 'minute',how = 'inner'),data3,on='minute',how='left')
#打标签（伪数据，设定47分和47分在吸烟）
result['is_smoke'] = 0
result.loc[result.minute.apply(lambda x : (x[-5:-3] =='47')or  (x[-5:-3] =='49')),'is_smoke'] = 1
result.query('is_smoke == 1')
result
```
